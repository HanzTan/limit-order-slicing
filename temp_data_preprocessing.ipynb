```python
import pandas as pd
import pickle
```

```python
# Temporary dataset made as shown in project brief
es_march = pd.read_csv("data/S&P500 Mini/ESH20.csv", header=None, index_col=0)
es_june = pd.read_csv("data/S&P500 Mini/ESM20.csv", header=None, index_col=0)
```

```python
march_df = es_march["2019.12.15":"2020.03.18"].copy()
june_df = es_june["2020.03.18":"2020.06.18"].copy()
es_df = pd.concat([march_df, june_df])
es_df.columns = ['o', 'h', 'l', 'c', 'v']
es_df.index.name = 't'

es_df
```

```python
es_df.to_csv("data/S&P500 Mini/ES_temp.csv")
```
